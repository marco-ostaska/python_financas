```python
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import yfinance as yf
```

```python
carteira = yf.download(["MGLU3.SA", "CASH3.SA", "GGBR4.SA", "CAML3.SA"], start="2019-01-01")['Adj Close'].dropna()

carteira.tail()
```

```python
(carteira / carteira.iloc[0] *100).plot(figsize=(15,5))
```

```python
retorno = (carteira / carteira.shift(1)) - 1
retorno.head()
```

```python
distribuicao = np.array([0.25, 0.25, 0.25, 0.25])
retorno_anual= retorno.mean() *250
print(round(np.dot(retorno_anual, distribuicao), 2) *100, "%")
```
