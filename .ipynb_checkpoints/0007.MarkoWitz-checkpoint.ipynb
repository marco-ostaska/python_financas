```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tickers=["BBDC4.SA", "GGBR4.SA", "BRAP4.SA", "CMIG4.SA", "CAML3.SA"]
dist=np.array([19.13,18.49, 1.39, 0.08, 0.97]) /100
ativos = yf.download(tickers, start="2001-01-01")['Adj Close'].dropna()
ativos.tail()
```

```python
(ativos / ativos.iloc[0] * 100).plot(figsize=(10,5))
```

```python
log_returns= np.log(ativos / ativos.shift(1))
log_returns.mean() * 250
```

```python
log_returns.cov() * 250
```

```python
log_returns.corr()
```

```python
numm_ativos = len(tickers)
numm_ativos
```

```python
dist = np.random.random(numm_ativos)
dist /= np.sum(dist)
dist
```

```python
np.sum(dist * log_returns.mean()) * 250
```

```python
np.dot(dist.T, np.dot(log_returns.cov() * 250, dist))
```

```python
np.sqrt(np.dot(dist.T, np.dot(log_returns.cov() * 250, dist)))
```

```python
retorno = []
volatividade = []

for _ in range(1000):
    dist = np.random.random(numm_ativos)
    dist /= np.sum(dist)
    retorno.append(np.dot(dist.T, np.dot(log_returns.cov() * 250, dist)))
    volatividade.append(np.sqrt(np.dot(dist.T, np.dot(log_returns.cov() * 250, dist))))

retorno = np.array(retorno)
volatividade = np.array(volatividade)
```

```python
portfolios = pd.DataFrame({'Retorno': retorno, 'Volatividade': volatividade})
portfolios.tail()
```

```python
portfolios.plot(x='risco', y='retorno', kind=)
```
